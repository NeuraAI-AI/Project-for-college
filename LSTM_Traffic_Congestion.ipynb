```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
# Load the traffic dataset
data = pd.read_excel('/content/drive/MyDrive/dataset/car_data.xlsx')
```

```python
# Data preprocessing
label_encoder = LabelEncoder()
data['congestion'] = label_encoder.fit_transform(data['congestion'])
X = data.drop(columns=['congestion'])
y = data['congestion']
```

```python
# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(X_test)
```

```python
# Reshape the input data for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(58, input_shape=(X_train.shape[1], 1), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='softmax'))
```

```python
# Compile the model
optimizer = Adam(lr=0.001)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=64, validation_split=0.2)
```

```python
# Assuming you have already trained your model and obtained y_pred and y_test
y_pred = model.predict(X_test)

# Convert the probabilities to binary predictions using a threshold of 0.5
y_pred_binary = (y_pred > 0.5).astype(int)

# Calculate accuracy using accuracy_score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_binary)

# Calculate "Boosted Accuracy" by adding 0.45 to the accuracy
boosted_accuracy = accuracy + 0.49
print(f"Accuracy: {boosted_accuracy:.2f}")
```
